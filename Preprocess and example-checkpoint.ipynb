```python
from scipy import io
import os
import numpy as np
from scipy.sparse.linalg import eigsh
import matplotlib.pyplot as plt
%matplotlib inline

dat0 = io.loadmat('6845348/natimg2800_M160825_MP027_2016-12-14.mat')
dat1 = io.loadmat('6845348/natimg2800_M161025_MP030_2017-05-29.mat')
dat2 = io.loadmat('6845348/natimg2800_M170604_MP031_2017-06-28.mat')
dat3 = io.loadmat('6845348/natimg2800_M170714_MP032_2017-08-07.mat')
dat4 = io.loadmat('6845348/natimg2800_M170714_MP032_2017-09-14.mat')
dat5 = io.loadmat('6845348/natimg2800_M170717_MP033_2017-08-20.mat')
dat6 = io.loadmat('6845348/natimg2800_M170717_MP034_2017-09-11.mat')

dats = [dat0,dat1,dat2,dat3,dat4,dat5,dat6]

resps = []
sponts = []
istims = []
nimgs = []

for i in range(0,7):
    resp = dats[i]['stim'][0]['resp'][0] # stim x neurons
    spont = dats[i]['stim'][0]['spont'][0] # timepts x neurons
    istim = (dats[i]['stim'][0]['istim'][0]).astype(np.int32) # stim ids 
    istim -= 1 # get out of MATLAB convention
    istim = istim[:,0]
    nimg = istim.max() # these are blank stims (exclude them)
    resp = resp[istim<nimg, :]
    istim = istim[istim<nimg]
    resps.append(resp)
    sponts.append(spont)
    istims.append(istim)
    nimgs.append(nimg)

# subtract spont (32D)
for i in range(0,7):
    mu = sponts[i].mean(axis=0)
    sd = sponts[i].std(axis=0) + 1e-6
    resp = (resps[i] - mu) / sd
    spont = (sponts[i] - mu) / sd
    sv,u = eigsh(spont.T @ spont, k=32)
    resp = resp - (resp @ u) @ u.T
    resps[i] = resp
    sponts[i] = spont

# mean center each neuron
for i in range(0,7):
    resps[i] -= resps[i].mean(axis=0)
```

Let's train a linear classifier on the data and test it (using a 85-15 train-test split). The fitting can take some time because there is a lot of data.

```python
from sklearn.model_selection import train_test_split

X_trains = []
X_tests = []
y_trains= []
y_tests = []

for i in range(0,7):
    X_train, X_test, y_train, y_test = train_test_split(resps_as_df[i].drop('label',axis=1), resps_as_df[i]['label'], test_size=0.15, random_state=0)
    X_trains.append(X_train)
    X_tests.append(X_test)
    y_trains.append(y_train)
    y_tests.append(y_test)
```

```python
for i in range(0,7):
    print(len(X_trains[i]))
```

```python
from sklearn.linear_model import RidgeClassifier

ridge = RidgeClassifier(random_state=0)
ridge.fit(X_trains[5],y_trains[5])
```

```python
preds = ridge.predict(X_tests[5])
```

```python
from sklearn.metrics import accuracy_score
print("Accuracy score: " + str(accuracy_score(preds,y_tests[5])))
```
